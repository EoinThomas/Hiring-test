# Excercise 3: Monthly searches for 3 airports

Plot the number of searches for 3 airport destinations.

The code is based on a groupby function using 2 indexes. The plot is generated directly by the groupby function and is serviceable for internal analysis as all , but some formatting was added to improve visual appeal. The code is compatible with any number/combination of airports.


```python
import pandas as pd
import calendar
```

```python
list_ports = ['AGP','MAD','BCN'] # code is compatible with any number of airports
list_ports.sort() # ensures the legend is accurate in plot
print list_ports
```

```python
data_searches_dest = pd.DataFrame() #initialise dataframe for data filtered by destination
searches_reader = pd.read_csv('searches.csv',sep = '^', usecols = ['Destination','Date'],chunksize = 100000) #loading the relevant columns

for chunk in searches_reader:
    data_searches_dest = data_searches_dest.append(chunk[chunk.Destination.isin(list_ports)]) #selecting relevant rows
    
del chunk,searches_reader
```

```python
dates = pd.DatetimeIndex(data_searches_dest.Date) #using pandas to automate the month extraction. Alternative: slice the strings.
data_searches_dest['Month'] = pd.np.transpose(dates.month) #Adding month info back into dataframe
```

```python
data_searches_dest.rename(columns={'Date': 'Searches'}, inplace=True) # Trick for formatting table/plot, data within is irrelevant
gby_month_dest = data_searches_dest.groupby(['Month','Destination']) # Groupby month and destination respectively
#gby_month_dest.count()  #Displays a table equivalent to a histogram, useful sanity check
```

```python
fig = gby_month_dest.count().unstack().plot() #Basic plot, axes and legend will need to be edited
fig.set_ylabel('Searches')
fig.set_xlabel('')
fig.set_title('Monthly searches per airport destination')
fig.legend(list_ports);
fig.set_xlim(1,12)
fig.set_xticks(range(1,13))
fig.set_xticklabels(calendar.month_name[1:13],rotation = 45);
```
