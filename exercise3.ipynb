# Excercise 3: Monthly searches for 3 airports

Plot the number of searches for 3 airport destinations

There seems to be some issues with the date parsing in pandas, I am using the latest stable release 0.14, which has some useful functionality recently added.

```python
import pandas as pd
import matplotlib.pyplot as matplot
```

```python
data_searches = pd.read_csv('searches.csv',sep = '^', usecols = ['Destination','Date']) #loading the relevant columns
data_searches_dest = data_searches[data_searches.Destination.isin(['BCN','MAD','AGP'])] #selecting relevant rows
del data_searches
```

```python
dates = pd.DatetimeIndex(data_searches_dest.Date) #using pandas to automate the month extraction.
data_searches_dest['Month'] = pd.np.transpose(dates.month) #Adding month info back into dataframe
```

```python
data_searches_dest.rename(columns={'Date': 'Searches'}, inplace=True) # Trick for formatting table/plot
gby_month_dest = data_searches_dest.groupby(['Month','Destination']) # Groupby month and destination, order matters
gby_month_dest.count()  #Displays a table equivalent to a histogram
```

```python
gby_month_dest.count().unstack().plot() #Basic plot, axes and legend will need to be edited
```
